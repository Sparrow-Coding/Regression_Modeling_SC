```python
def Regression_Models(X,y):
    import numpy as np
    from sklearn.model_selection import train_test_split
    from sklearn.model_selection import cross_val_score
    from matplotlib import pyplot as plt 
    from sklearn.linear_model import LinearRegression
    from sklearn.linear_model import RidgeCV
    from sklearn.linear_model import LassoCV
    from sklearn.svm import SVR
    from sklearn.tree import DecisionTreeRegressor
    from sklearn.ensemble import RandomForestRegressor, BaggingRegressor, AdaBoostRegressor
    from sklearn.neighbors import KNeighborsRegressor
    from sklearn.metrics import mean_squared_error
    from sklearn.model_selection import GridSearchCV
    
    param_grid = {"alpha": np.linspace(0.00001, 1, 20)}
    X= X
    y= y
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=16)

    # Linear Regression
    lr = LinearRegression()
    lr.fit(X_train,y_train)
    lr.score(X_train,y_train)
    lr.score(X_test,y_test)
    np.mean(cross_val_score(lr, X, y, cv=5))
    print('Linear Regression Train Score:', lr.score(X_train,y_train),'|','Linear Regression Test Score:',lr.score(X_test,y_test))
    
   # Lasso Regression
    lasso = LassoCV()
    lasso.fit(X_train, y_train)
    lasso.score(X_train, y_train)
    lasso.score(X_test, y_test)
    print('Lasso Regression Train Score:', lasso.score(X_train,y_train),'|','Lasso Regression Test Score:',lasso.score(X_test,y_test))
    
    # Decision Tree
    dt = DecisionTreeRegressor()
    dt.fit(X_train,y_train)
    dt.score(X_train, y_train)
    dt.score(X_test,y_test)
    print('DecisionTree Regression Train Score:', dt.score(X_train,y_train),'|','DecisionTree Regression Test Score:',dt.score(X_test,y_test))
    
    # Random Forest
    rf = RandomForestRegressor()
    rf.fit(X_train,y_train)
    rf.score(X_train, y_train)
    rf.score(X_test,y_test)
    print('RandomForest Regression Train Score:', rf.score(X_train,y_train),'|','RandomForest Regression Test Score:',rf.score(X_test,y_test))
    
    # KNeighbors Regressor
    knn = KNeighborsRegressor()
    knn.fit(X_train,y_train)
    knn.score(X_train, y_train)
    knn.score(X_test, y_test)
    print('KNeighbors Regression Train Score:', knn.score(X_train,y_train),'|','KNeighbors Regression Test Score:',knn.score(X_test,y_test))
    
    # Bagging Regressor
    br = BaggingRegressor()
    br.fit(X_train,y_train)
    br.score(X_train, y_train)
    br.score(X_test, y_test)
    print('Bagging Regression Train Score:', br.score(X_train,y_train),'|','Bagging Regression Test Score:',br.score(X_test,y_test))
    
    # Ada Boost Regressor
    abr = AdaBoostRegressor()
    abr.fit(X_train,y_train)
    abr.score(X_train, y_train)
    abr.score(X_test, y_test)
    print('AdaBoost Regression Train Score:', abr.score(X_train,y_train),'|','AdaBoost Regression Test Score:',abr.score(X_test,y_test))
    
    # Support Vector Regressor
    sv = SVR()
    sv.fit(X_train,y_train)
    sv.score(X_train, y_train)
    sv.score(X_test, y_test)
    print('Support Vector Regression Train Score:', sv.score(X_train,y_train),'|','Support Vector Regression Test Score:',sv.score(X_test,y_test))
    
  
     # Ridge Regression
    alphas = [0.1, 1.0, 10.0, 100.0, 1000.0, 10000.0]
    ridge_test_scores = []
    ridge_train_scores = []
    for alpha in alphas:
        ridge = RidgeCV(alphas=alpha)
        ridge.fit(X_train,y_train)
        test_score = ridge.score(X_test,y_test)
        train_score = ridge.score(X_train,y_train)
        ridge_test_scores.append(test_score)
        ridge_train_scores.append(train_score) 
    print(f"ridge train scores are:{ridge_train_scores}")
    print(f"ridge test scores are:{ridge_test_scores}")
   
```
